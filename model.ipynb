### Stats0 process

```python
df_stats0 = pd.read_csv('data/Premier-League/Premier-League_stats_0.csv', header=1, index_col=0)
df_stats0.head()
```

```python
# Remove useless columns
df_stats0.drop(columns=['Affluence', 'Meilleur buteur de l\'équipe', 'Gardien de but', 'Notes'], inplace=True)

# Drop column that are already in another dataset
df_stats0.drop(columns=['xG', 'xGA', 'xGD', 'xGD/90'], inplace=True)

df_stats0.head()
```

```python
# Convert the 5 last matches column into usable data
df_stats0[['dernier_1', 'dernier_2', 'dernier_3', 'dernier_4', 'dernier_5']] = df_stats0['5 derniers'].str.split(' ', expand=True)
df_stats0[['dernier_1', 'dernier_2', 'dernier_3', 'dernier_4', 'dernier_5']] = df_stats0[['dernier_1', 'dernier_2', 'dernier_3', 'dernier_4', 'dernier_5']].replace({'D': -1, 'N': 0, 'V': 1})

df_stats0.drop(columns=['5 derniers'], inplace=True)

df_stats0.head()
```
