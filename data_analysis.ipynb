## Process features

```python
regular_season_df = pd.read_csv("data/Regular season Table_0.csv",header=1, index_col=0)
regular_season_df
```

```python
schedule_df
```

```python
merged_df = merge_statistics_to_match_schedule(df_schedule=schedule_df, df_statistics=regular_season_df)
merged_df.head()
```

```python
X = get_features(merged_df)

y_dom = get_home_target(merged_df)
y_ext = get_ext_target(merged_df)

print(f"X.shape: {X.shape}")
print(f"y_dom.shape: {y_dom.shape}")
print(f"y_ext.shape: {y_ext.shape}")
```

```python
print(f"number na in X: {X.isna().sum().sum()}")
print(f"number na in y_dom: {y_dom.isna().sum()}")
print(f"number na in y_ext: {y_ext.isna().sum()}")
```

```python
import sklearn
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import train_test_split

X_train, X_test, y_dom_train, y_dom_test = train_test_split(X, y_dom, test_size=0.2, random_state=42)

print(f"X_train.shape: {X_train.shape}")
print(f"y_train.shape: {y_dom_train.shape}")
print(f"X_test.shape: {X_test.shape}")
print(f"y_test.shape: {y_dom_test.shape}")
```

```python
gb_dom = GradientBoostingClassifier(n_estimators=100, learning_rate=0.1,
                                 max_depth=1, random_state=42).fit(X_train, y_dom_train)
```

```python
gb_dom.score(X_test, y_dom_test)
```

```python
X_train, X_test, y_ext_train, y_ext_test = train_test_split(X, y_ext, test_size=0.2, random_state=42)
```

```python
gb_ext = GradientBoostingClassifier(n_estimators=100, learning_rate=0.1,
                                 max_depth=1, random_state=42).fit(X_train, y_ext_train)
```

```python
gb_ext.score(X_test, y_ext_test)
```
